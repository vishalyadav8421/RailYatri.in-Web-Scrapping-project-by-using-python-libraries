### Now i have to Request with tose website to give me their data so i use below codes

```python
page = requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
soup.prettify
```

```python
page
```

### As i got the Response 200 it means i got the acess from website to scrap data from it

### After geting response i will paste the class code to extract data from that class only

```python
train_detail=soup.find_all(class_="table table-condensed")
print(train_detail)
print(len(train_detail))
```

# DATA CLEANING IS OUR MAIN AIM FOR ACCURATE DATA EXTRACTION 

###### I WILL CREATE SOME LIST AND DOING FURTHER PROCESS OF DATA CLEANING 

```python
list1=[]
for i in range(0,len(train_detail)):
    list1.append(train_detail[i].get_text())
list1
```

###### AS I GOT MY DATA WAS NESTED IN \N\N\N\N TAGS SO I USE SPLIT FUNCTION TO REMOVE IT

```python
list2=[]
for i in range(0,len(list1)):
    list2.append(list1[i].split("\n"))
print(list2[0])
```

###### AFTER REMOVING \N \N  I GOT ON THERE PLACE SOME COMMAS (  "  ) VALUES NOT AGAIN I USE SPLIT FUNCTION TO REMOVE IT

```python
list3=[]
for i in range(0,len(list2)):
    for n in range(0,len(list2[i])):
        list3.append(list2[i][n].split(','))
print(list3)
```

######  I WILL CREATE SOME EMPTY LIST OF VALUES BY SPLITING IT WITH COMMA ( , )

```python
for i in range(0,len(list3)):
    for n in range(0,len(list3[i])):
        if list3[i][n]=='':
            list3[i].remove('')
print(len(list3))
print(list3)
```

###### NOW REMOVING ALL BLANK LIST BY USING LENGTH LEN( ) FUNCTION BY REVERSE INDEXING

```python
for i in range(len(list3)-1,-1,-1):
    if len(list3[i])==0:
        list3.remove(list3[i])
list3
```

###### FOR GETTING OUR CORRECT DATA NOW I WILL REMOVE SOME UNWANTED DATA FROM IT 

```python
list3[0:6]
```

```python
detail=[]
detail.append(list3[0:6])
detail
```

######  Now removing the Detail list all values in our main list before passing it in a loop

```python
list3[0:6]=[]     # for removing values from index 0 to 6
list3
len(list3)
```

###### Empty list for appending our different data

```python
Train_Number=[]
Train_Name=[]
Train_Type=[]
Run_Days=[]
Departure_time=[]
Arrival_Time=[]
```

### APPENDING ALL VALUES IN EMPTY LIST BY USING JUMPER(HOPPING) IN INDEX METHOD 

```python
#append values in lisst one by one by using hopping or jumper in main list

for i in range(0,len(list3),6):                   #FOR INEDX 0TH TO JUMP AT 6TH INDEX
    Train_Number.append(list3[i])
    
for i in range(1,len(list3),6):                  #FOR INEDX 1TH TO JUMP AT 6TH INDEX
    Train_Name.append(list3[i])
    
for i in range(2,len(list3),6):                  #FOR INEDX 2TH TO JUMP AT 6TH INDEX
    Train_Type.append(list3[i])
    
for i in range(3,len(list3),6):                  #FOR INEDX 3TH TO JUMP AT 6TH INDEX
    Run_Days.append(list3[i])
    
for i in range(4,len(list3),6):                  #FOR INEDX 4TH TO JUMP AT 6TH INDEX
    Departure_time.append(list3[i])
    
for i in range(5,len(list3),6):                  #FOR INEDX 05TH TO JUMP AT 6TH INDEX
    Arrival_Time.append(list3[i])
```
